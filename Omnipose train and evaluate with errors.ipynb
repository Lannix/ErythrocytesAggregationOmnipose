## Анализ видео с агрегацией эритроцитов

### Необходимые функции

```python
import cv2
import os
import subprocess
import sys
from tqdm import tqdm, trange
import matplotlib.pyplot as plt
from skimage.measure import regionprops
import numpy as np
import shutil
from skimage.io import imread
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from typing import List, Tuple, Dict
from skimage.measure._regionprops import RegionProperties
from scipy import stats


def save_video_frames(video_path: str, result_path: str, n_frames: int):
    """
    Сохраняет 'n_frames' снимков из видео в указанную папку.

    Параметры:
    - video_path: str, путь к видеофайлу.
    - result_path: str, путь к папке, куда сохранять снимки.
    - n_frames: int, количество снимков для сохранения.
    """
    # Проверяем, существует ли папка, если нет — создаём
    if not os.path.exists(result_path):
        os.makedirs(result_path)
    
    # Открываем видеофайл
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("Ошибка: не удалось открыть видеофайл.")
        return
    
    # Получаем общее количество кадров в видео
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    
    # Определяем шаг для выборки кадров
    step = max(total_frames // n_frames, 1)  # Чтобы избежать деления на 0
    
    saved_frames = 0
    frame_count = 0
        
    while saved_frames < n_frames and frame_count < total_frames:
        ret, frame = cap.read()
        if not ret:
            break
        
        # Сохраняем кадр, если он попадает в выборку
        if frame_count % step == 0 or saved_frames == 0:
            frame_name = os.path.join(result_path, f"frame_{frame_count}.png")
            cv2.imwrite(frame_name, frame)
            saved_frames += 1
        
        frame_count += 1
    
    fps = cap.get(cv2.CAP_PROP_FPS)
    cap.release()
    return fps


def run_omnipose(images_dir: str, model_path: str):
    """
    Запускает Omnipose с указанными параметрами, выводя лог в реальном времени
    (адаптировано для Jupyter Notebook).

    Args:
        images_dir (str): Путь к папке с изображениями (например, "C:\\data\\images")
        model_path (str): Путь к файлу модели (например, "C:\\models\\omnipose_model")
    """
    # Проверяем существование путей
    if not os.path.isdir(images_dir):
        raise ValueError(f"Директория с изображениями не найдена: {images_dir}")
    if not os.path.isfile(model_path):
        raise ValueError(f"Файл модели не найден: {model_path}")

    # --- Улучшение: Использование `sys.executable` для надежности ---
    # Это гарантирует, что мы используем python-интерпретатор из того же окружения,
    # где запущен Jupyter Notebook, что решает 99% проблем с PATH и venv.
    cmd = [
        sys.executable, "-m", "omnipose",
        "--dir", images_dir,
        "--pretrained_model", model_path,
        "--no_npy",
        "--save_png",
        "--save_outlines",
        "--nchan", "2",
        "--all_channels",
        "--channel_axis", "2",
        "--diameter", "0",
        "--nclasses", "3"
    ]

    print("--- Запуск Omnipose ---")
    print(f"Команда: {' '.join(cmd)}")
    
    # --- Основное изменение для Jupyter ---
    # Используем Popen для запуска процесса в фоновом режиме и получения доступа к его выводу
    try:
        process = subprocess.Popen(
            cmd,
            stdout=subprocess.PIPE,          # Перехватываем стандартный вывод
            stderr=subprocess.STDOUT,         # Перенаправляем ошибки в тот же поток, что и вывод
            text=True,                        # Декодируем вывод в текст (str)
            encoding='utf-8',                 # Явно указываем кодировку
            bufsize=1                         # Включаем построчную буферизацию
        )

        # Читаем вывод строка за строкой, пока процесс работает
        if process.stdout:
            for line in iter(process.stdout.readline, ''):
                print(line, end='')  # Печатаем строку в ячейку вывода
                sys.stdout.flush()   # !!! Ключевой момент: принудительно сбрасываем буфер вывода

        # Ждем завершения процесса и получаем его код возврата
        return_code = process.wait()

        if return_code == 0:
            print("\n--- Omnipose успешно выполнен! ---")
        else:
            # Сообщение об ошибке уже будет выведено в цикле выше
            print(f"\n--- Ошибка при выполнении Omnipose (код возврата: {return_code}) ---")
            raise subprocess.CalledProcessError(return_code, cmd)

    except FileNotFoundError:
        print("\nОшибка: Не удалось запустить omnipose.")
        print("Убедитесь, что omnipose установлен в текущем Python окружении (`pip install omnipose`).")
        raise
    except subprocess.CalledProcessError as e:
        # Это исключение будет перехвачено и обработано выше, но на всякий случай
        print(f"Процесс завершился с ошибкой: {e}")
        raise


def train_cell_cluster_model(
    mask_dir: str, 
    features_to_use: List[str] = ['area', 'solidity', 'eccentricity']
) -> Tuple[StandardScaler, KMeans, Dict[int, str]]:
    """
    Анализирует все объекты во всех масках и обучает модель K-Means для их разделения на 2 класса.

    Параметры:
    - mask_dir: Путь к папке с масками. Там же должны быть соответствующие изображения.
    - features_to_use: Список свойств из regionprops для использования в кластеризации.
      'area' обязательна для определения, какой кластер является агрегатами.

    Возвращает:
    - scaler: Обученный StandardScaler для нормализации данных.
    - kmeans_model: Обученная модель KMeans.
    - label_map: Словарь, сопоставляющий метку кластера (0 или 1) с названием ('single_cell' или 'aggregate').
    """
    if 'area' not in features_to_use:
        raise ValueError("Список 'features_to_use' должен содержать 'area' для определения кластеров.")

    try:
        mask_files = [f for f in os.listdir(mask_dir) if f.endswith('_cp_masks.png')]
        image_files = [f for f in os.listdir(mask_dir) if f.endswith('_c.png')]
    except FileNotFoundError:
        raise FileNotFoundError(f"Директория не найдена: {mask_dir}")
        
    if not mask_files:
        raise FileNotFoundError(f"Не найдены файлы масок с '_cp_masks.png' в директории: {mask_dir}")
        
    all_features = []
    print(f"Сбор данных для обучения с {len(mask_files)} масок...")
    
    for mask_file in tqdm(mask_files, desc="Сбор признаков"):
        mask_path = os.path.join(mask_dir, mask_file)
        mask = imread(mask_path).astype(np.uint16)
        image_file = mask_file.replace('_cp_masks.png', '.png')
        image_path = os.path.join(mask_dir, image_file)
        image = imread(image_path, as_gray=True).astype(np.uint8)
        regions = regionprops(mask, intensity_image=image)
        
        for region in regions:
            # Собираем только нужные признаки для каждого объекта
            current_features = [getattr(region, feature) for feature in features_to_use]
            all_features.append(current_features)
            
    if not all_features:
        raise ValueError("В масках не найдено ни одного объекта для обучения модели.")

    X = np.array(all_features)
    print(f"Обучение модели на {X.shape[0]} объектах...")

    # 1. Нормализация данных: K-Means чувствителен к масштабу признаков.
    # Площадь может быть 1000+, а солидность < 1. Их нужно привести к одному масштабу.
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # 2. Обучение модели K-Means для поиска 2 кластеров
    # n_init='auto' для подавления Future-предупреждения в новых версиях sklearn
    kmeans_model = KMeans(n_clusters=2, random_state=42, n_init='auto')
    kmeans_model.fit(X_scaled)
    labels = kmeans_model.labels_

    # 3. Определяем, какой кластер - "агрегаты", а какой - "одиночные клетки"
    # Логика: у кластера агрегатов средняя площадь будет больше.
    area_index = features_to_use.index('area')
    mean_area_cluster_0 = X[labels == 0, area_index].mean()
    mean_area_cluster_1 = X[labels == 1, area_index].mean()

    if mean_area_cluster_0 > mean_area_cluster_1:
        aggregate_label = 0
        single_cell_label = 1
    else:
        aggregate_label = 1
        single_cell_label = 0
        
    label_map = {
        aggregate_label: 'aggregate',
        single_cell_label: 'single_cell'
    }
    
    print("Обучение завершено.")
    print(f"Кластер {aggregate_label} определен как 'aggregate' (ср. площадь: {max(mean_area_cluster_0, mean_area_cluster_1):.2f})")
    print(f"Кластер {single_cell_label} определен как 'single_cell' (ср. площадь: {min(mean_area_cluster_0, mean_area_cluster_1):.2f})")
    
    return scaler, kmeans_model, label_map


def separate_regions_by_class(
    regions: List[RegionProperties],
    scaler: StandardScaler,
    model: KMeans,
    label_map: Dict[int, str],
    features_to_use: List[str] = ['area', 'solidity', 'eccentricity']
) -> Tuple[List[RegionProperties], List[RegionProperties]]:
    """
    Классифицирует и фильтрует список объектов regionprops на две группы.

    Использует предварительно обученные scaler и модель KMeans для предсказания
    класса каждого объекта ('single_cell' или 'aggregate').

    Параметры:
    - regions (List[RegionProperties]): Список объектов, полученный из skimage.measure.regionprops.
    - scaler (StandardScaler): Обученный нормализатор для признаков.
    - model (KMeans): Обученная модель кластеризации.
    - label_map (Dict[int, str]): Словарь для сопоставления ID кластера с его именем
      (например, {0: 'aggregate', 1: 'single_cell'}).
    - features_to_use (List[str]): Список названий признаков, которые использовались
      для обучения модели. Порядок должен быть тем же.

    Возвращает:
    - Tuple[List[RegionProperties], List[RegionProperties]]: Кортеж из двух списков:
        1. cell_regions: Список объектов, классифицированных как 'single_cell'.
        2. agg_regions: Список объектов, классифицированных как 'aggregate'.
    """
    # Инициализируем пустые списки для результатов
    cell_regions: List[RegionProperties] = []
    agg_regions: List[RegionProperties] = []

    # Перебираем каждый обнаруженный объект
    for region in regions:
        try:
            # 1. Извлекаем необходимые признаки для объекта
            current_features = [getattr(region, feature) for feature in features_to_use]
            
            # 2. Преобразуем в 2D-массив для модели
            features_array = np.array(current_features).reshape(1, -1)
            
            # 3. Нормализуем признаки с помощью того же scaler'а, что и при обучении
            features_scaled = scaler.transform(features_array)
            
            # 4. Предсказываем ID кластера
            predicted_label_id = model.predict(features_scaled)[0]
            
            # 5. Получаем имя класса из словаря
            class_name = label_map[predicted_label_id]
            
            # 6. Добавляем объект в соответствующий список
            if class_name == 'single_cell':
                cell_regions.append(region)
            elif class_name == 'aggregate':
                agg_regions.append(region)
                
        except AttributeError as e:
            print(f"Предупреждение: не удалось извлечь признак для объекта. {e}. Объект пропущен.")
        except Exception as e:
            print(f"Предупреждение: произошла ошибка при классификации объекта. {e}. Объект пропущен.")
            
    return cell_regions, agg_regions


def analyze_cell_masks(mask_dir: str, fps: float, px_per_um: float = 3.8, cluster = True):
    """
    Анализирует последовательность масок клеток и их агрегатов и строит графики зависимости количества и площади от времени.
    
    Предполагается, что входные файлы масок являются мечеными изображениями (labeled images), 
    где каждый отдельный объект (клетка) имеет уникальное целочисленное значение.
    
    Параметры:
    - mask_dir: Путь к папке с масками (только файлы с '_cp_masks.png').
    - fps: Частота кадров для расчета времени кадра по его номеру.
    - px_per_um: Количество пикселей в одном микрометре (по умолчанию 3.8).
    
    Возвращает:
    - time_points: Временные точки в секундах.
    - aggregates_counts: Количество агрегатов клеток на каждом кадре.
    - avg_areas_um2: Средняя площадь объектов на каждом кадре (в мкм²).
    - avg_counts_in_aggregates: Среднее количество клеток в агрегатах ((начальное кол-во клеток - текущее количество клеток) / текущее кол-во агрегатов).
    - cluster: использовать ли KMeans для классификации клеток и их агрегатов. Позволяет получить кинетику именно агрегации, свободные клетки не вносят вклада.
    """
    
    # Получаем и фильтруем список файлов масок
    try:
        mask_files = sorted(
            [f for f in os.listdir(mask_dir) if f.endswith('_cp_masks.png')], 
            key=lambda x: int(x.split('_')[1])
        )
    except IndexError:
        raise ValueError("Формат имен файлов не соответствует ожидаемому 'prefix_framenumber_...'. Не удалось извлечь номер кадра.")
        
    if not mask_files:
        raise FileNotFoundError(f"Не найдены файлы масок с '_cp_masks.png' в директории: {mask_dir}")
    
    time_points = []
    aggregates_counts = []
    avg_areas_um2 = []
    avg_counts_in_aggregates = []
    areas_sems = []
    counts_in_aggregates_sems = []

    start_regions_count = 0 # количество клеток в самом начале на момент полной дезагрегации
    px_to_um2_factor = px_per_um ** 2

    features = ['area', 'axis_major_length', 'solidity', 'eccentricity', 'intensity_mean', 'intensity_std']

    if cluster:
        print(f"Обучаем KMeans для разделения клеток и агрегатов...")
        
        scaler, model, label_map = train_cell_cluster_model(
            mask_dir=mask_dir,
            features_to_use=features
        )
    
    print(f"Анализируем {len(mask_files)} масок...")
    
    for i, mask_file in enumerate(tqdm(mask_files, desc="Анализ масок")):
        # Загружаем маску
        mask_path = os.path.join(mask_dir, mask_file)
        # Используем imread из scikit-image, он лучше подходит для работы с мечеными масками
        mask = imread(mask_path)        
        image_file = mask_file.replace('_cp_masks.png', '.png')
        image_path = os.path.join(mask_dir, image_file)
        image = imread(image_path, as_gray=True).astype(np.uint8)
        
        # Номер кадра из имени файла
        frame_number = int(mask_file.split('_')[1])
        
        labeled_mask = mask.astype(np.uint16)
        regions = regionprops(labeled_mask, intensity_image=image)
        num_regions = len(regions)

        # Выделяем кластеры
        if cluster:
            cell_regions, agg_regions = separate_regions_by_class(regions, scaler, model, label_map, features_to_use=features)
        else:
            agg_regions = regions

        num_aggregates = len(agg_regions)
        num_cells = num_regions - num_aggregates
        
        # Конвертируем площади из пикселей в мкм²
        agg_areas = [r.area / px_to_um2_factor for r in agg_regions]
        
        # Собираем статистику
        time_points.append(frame_number / fps)
        aggregates_counts.append(num_aggregates)

        # предполагаем, что в начале все клетки находятся в полностью дезагрегированном состоянии
        if i == 0:
            start_regions_count = num_regions
        
        if num_aggregates > 0:
            aggregates_area = sum(agg_areas)
            areas_sems.append(stats.sem(agg_areas))
            avg_areas_um2.append(aggregates_area / num_aggregates)
            # Эта метрика показывает, во сколько раз уменьшилось количество объектов (например, из-за слияния)
            avg_counts_in_aggregates.append((start_regions_count - num_cells) / num_aggregates)
            # Рассчитаем стандартную ошибку количества клеток в агрегатах, считая, что количество клеток прямо пропорционально площади агрегатов
            counts_in_aggregates_sems.append(areas_sems[-1] / avg_areas_um2[-1] * avg_counts_in_aggregates[-1])
        else:
            # Если на кадре нет объектов
            avg_areas_um2.append(0)
            avg_counts_in_aggregates.append(0) # Если начинали с 0, то 0. Если нет - то бесконечность

    return time_points, aggregates_counts, avg_areas_um2, areas_sems, avg_counts_in_aggregates, counts_in_aggregates_sems


def get_video_fps(video_path):
    """
    Возвращает FPS (кадры в секунду) видеофайла.

    Параметры:
    - video_path (str): Путь к видеофайлу.

    Возвращает:
    - float: FPS видео.
    - None: Если видео не удалось открыть.
    """
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Ошибка: не удалось открыть видео {video_path}!")
        return None
    
    fps = cap.get(cv2.CAP_PROP_FPS)  # Получаем FPS
    cap.release()  # Закрываем видео
    
    return fps


def analyse_video_aggregation(video_path, n_frames, model_path, output_frames_path, omnipose=True, cluster = True):
    tmp_path = os.path.join(os.getcwd(), 'output_frames')
    
    if omnipose:
        fps = save_video_frames(video_path, tmp_path, n_frames)
        run_omnipose(tmp_path, model_path)
    else:
        fps = get_video_fps(video_path)
        

    time_points, aggregates_counts, avg_areas_um2, areas_sems, avg_counts_in_aggregates, counts_in_aggregates_sems = analyze_cell_masks(tmp_path, fps, px_per_um=3.8, cluster=cluster)

    # Строим графики

    # График количества клеток
    plt.plot(time_points, aggregates_counts, marker='o', linestyle='--', markersize=4)
    plt.title('Количество агрегатов')
    plt.xlabel('Время, с')
    plt.ylabel('Количество')
    plt.tight_layout()
    plt.show()
    
    # --- График средней площади с закрашенной областью ошибки ---
    plt.figure(figsize=(8, 5))
    # Рисуем основную линию
    plt.plot(time_points, avg_areas_um2, marker='o', linestyle='--', label='Средняя площадь')
    # Закрашиваем область ошибки
    plt.fill_between(
        time_points,
        np.array(avg_areas_um2) - np.array(areas_sems),        # Нижняя граница
        np.array(avg_areas_um2) + np.array(areas_sems),        # Верхняя граница
        color='blue',                      # Цвет области
        alpha=0.2,                         # Прозрачность
        label='Диапазон ст. ошибки'
    )
    plt.title('Средняя площадь агрегатов')
    plt.xlabel('Время, с')
    plt.ylabel('Площадь (мкм²)')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.tight_layout()
    plt.show()
    
    
    # --- График среднего кол-ва клеток с закрашенной областью ошибки ---
    plt.figure(figsize=(8, 5))
    # Рисуем основную линию
    plt.plot(time_points, avg_counts_in_aggregates, marker='o', linestyle='--', color='green', label='Среднее кол-во клеток')
    # Закрашиваем область ошибки
    plt.fill_between(
        time_points,
        np.array(avg_counts_in_aggregates) - np.array(counts_in_aggregates_sems), # Нижняя граница
        np.array(avg_counts_in_aggregates) + np.array(counts_in_aggregates_sems), # Верхняя граница
        color='green',
        alpha=0.2,
        label='Диапазон ст. ошибки'
    )
    plt.title('Среднее количество клеток в агрегатах')
    plt.xlabel('Время, с')
    plt.ylabel('Количество клеток в агрегате')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.tight_layout()
    plt.show()


def delete_folder(folder_path):
    """
    Безопасно удаляет папку со всем её содержимым.
    
    Параметры:
    - folder_path: str - путь к папке, которую нужно удалить
    
    Возвращает:
    - bool: True если удаление прошло успешно, False в случае ошибки
    """
    try:
        # Проверяем, существует ли папка
        if not os.path.exists(folder_path):
            print(f"Папка {folder_path} не существует")
            return False
            
        # Проверяем, что это действительно папка
        if not os.path.isdir(folder_path):
            print(f"{folder_path} не является папкой")
            return False
            
        # Удаляем папку со всем содержимым
        shutil.rmtree(folder_path)
        print(f"Папка {folder_path} успешно удалена")
        return True
        
    except PermissionError:
        print(f"Ошибка: нет прав для удаления папки {folder_path}")
    except Exception as e:
        print(f"Ошибка при удалении папки {folder_path}: {str(e)}")
    
    return False
```

### Функция для визуализации

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.measure import find_contours, regionprops
from matplotlib.lines import Line2D
import pandas as pd
from tqdm import tqdm
from typing import Optional, List

# Используем display для красивого вывода в Jupyter
from IPython.display import display

def visualize_segmentation_results(
    image_dir: str,
    num_images: int = 5,
    features_to_use: List[str] = ['area', 'axis_major_length', 'solidity', 'eccentricity', 'intensity_mean', 'intensity_std']
):
    """
    Визуализирует результаты сегментации, отображая классы объектов цветом.

    Функция сначала обучает модель для классификации объектов на "одиночные клетки"
    и "агрегаты". Затем выбирает 'num_images' кадров и для каждого отображает:
    1. Верхний ряд: оригинальное изображение.
    2. Нижний ряд: оригинальное изображение с наложенными контурами, где цвет
       контура соответствует классу объекта.

    Параметры:
    - image_dir: Путь к папке с оригинальными изображениями и масками.
    - num_images: Количество изображений для отображения (по умолчанию 5).
    - features_to_use: Признаки для обучения модели кластеризации.
    """
    # 1. Обучаем модель для классификации объектов
    print("--- Этап 1: Обучение модели кластеризации ---")
    try:
        scaler, kmeans_model, label_map = train_cell_cluster_model(
            mask_dir=image_dir, 
            features_to_use=features_to_use
        )
    except (FileNotFoundError, ValueError, IndexError) as e:
        print(f"Ошибка при обучении модели: {e}")
        print("Визуализация будет прервана.")
        return

    # Карта цветов для классов
    color_map = {
        'single_cell': 'lime',
        'aggregate': 'red'
    }
    
    print("\n--- Этап 2: Визуализация результатов ---")
    
    # 2. Получаем и сортируем список файлов масок
    try:
        mask_files = sorted(
            [f for f in os.listdir(image_dir) if f.endswith('_cp_masks.png')], 
            key=lambda x: int(x.split('_')[1])
        )
    except (FileNotFoundError, IndexError):
        print(f"Ошибка: не удалось найти или прочитать файлы масок в '{image_dir}'.")
        return

    if not mask_files:
        print(f"В папке '{image_dir}' не найдено файлов с масками ('_cp_masks.png').")
        return

    # 3. Выбираем N файлов через равные промежутки
    total_files = len(mask_files)
    if total_files < num_images:
        print(f"Найдено только {total_files} файлов, отображаем все.")
        selected_indices = np.arange(total_files)
    else:
        selected_indices = np.linspace(0, total_files - 1, num=num_images, dtype=int)
    
    selected_mask_files = [mask_files[i] for i in selected_indices]

    # 4. Создаем сетку для отображения изображений
    fig, axes = plt.subplots(2, num_images, figsize=(num_images * 4, 8.5))
    if num_images == 1:
        axes = np.array(axes).reshape(2, 1)

    print(f"Отображение {len(selected_mask_files)} кадров...")

    for i, mask_filename in enumerate(tqdm(selected_mask_files)):
        # 5. Формируем пути к файлам
        image_filename = mask_filename.replace('_cp_masks.png', '.png')
        frame_number = mask_filename.split('_')[1]
        
        image_path = os.path.join(image_dir, image_filename)
        mask_path = os.path.join(image_dir, mask_filename)
        
        try:
            image = imread(image_path)
            mask = imread(mask_path)
        except FileNotFoundError as e:
            print(f"Пропущен кадр {frame_number}: файл не найден. {e}")
            axes[0, i].axis('off'); axes[1, i].axis('off')
            continue

        # --- Отображение в верхнем ряду (Оригинал) ---
        ax_top = axes[0, i]
        ax_top.imshow(image, cmap='gray')
        ax_top.set_title(f'Кадр {frame_number}')
        ax_top.axis('off')

        # --- Отображение в нижнем ряду (Контуры с классами) ---
        ax_bottom = axes[1, i]
        ax_bottom.imshow(image, cmap='gray')
        ax_bottom.axis('off')

        # Находим все объекты на маске
        intensity_image = imread(image_path, as_gray=True).astype(np.uint8)
        regions = regionprops(mask, intensity_image=intensity_image)
        ax_bottom.set_title(f'Классы ({len(regions)} объектов)')

        # Перебираем каждый объект, классифицируем и отрисовываем
        for region in regions:
            # a) Извлекаем признаки для текущего объекта
            current_features = [getattr(region, feature) for feature in features_to_use]
            features_array = np.array(current_features).reshape(1, -1)
            
            # b) Нормализуем признаки с помощью обученного scaler'а
            features_scaled = scaler.transform(features_array)
            
            # c) Предсказываем класс объекта
            predicted_label_id = kmeans_model.predict(features_scaled)[0]
            class_name = label_map[predicted_label_id]
            
            # d) Получаем цвет для этого класса
            contour_color = color_map.get(class_name, 'white') # Белый по умолчанию
            
            # e) Находим контур только для текущего объекта и отрисовываем его
            # Создаем временную маску только с одним этим объектом
            mini_mask = np.zeros_like(mask, dtype=np.uint16)
            mini_mask[mask == region.label] = 1
            contours = find_contours(mini_mask, 0.5)

            for contour in contours:
                ax_bottom.plot(contour[:, 1], contour[:, 0], linewidth=1, color=contour_color)

    # 6. Добавляем легенду
    legend_elements = [
        Line2D([0], [0], color=color, lw=2, label=label.replace('_', ' ').title())
        for label, color in color_map.items()
    ]
    fig.legend(handles=legend_elements, loc='lower center', ncol=len(color_map), bbox_to_anchor=(0.5, 0.02))

    plt.tight_layout(rect=[0, 0.08, 1, 1]) # Оставляем место для легенды внизу
    plt.show()


def analyze_and_save_properties(
    mask_dir: str, 
    output_path: str,
    pixel_size_um: Optional[float] = None,
    properties_to_measure: List[str] = ['area', 'axis_major_length', 'solidity', 'eccentricity', 'mean_intensity']
) -> pd.DataFrame:
    """
    Анализирует объекты на масках, собирает их свойства и сохраняет в Excel.

    Функция итерируется по всем файлам масок ('_cp_masks.png'), находит объекты,
    извлекает для них морфологические и интенсивностные характеристики,
    собирает все данные в pandas DataFrame, выводит заголовок в консоль и
    сохраняет полный набор данных в .xlsx файл.

    Параметры:
    - mask_dir (str): Путь к директории, содержащей файлы масок и соответствующие им изображения.
    - output_path (str): Путь для сохранения итогового файла .xlsx.
    - pixel_size_um (Optional[float]): Размер одного пикселя в микрометрах. Если указан,
      добавляет в таблицу колонки с размерами в мкм.
    - properties_to_measure (List[str]): Список свойств для извлечения через regionprops.
      'intensity_std' вычисляется отдельно.

    Возвращает:
    - pd.DataFrame: DataFrame со всеми собранными данными.
    """
    # Проверяем, установлен ли openpyxl, необходимый для записи .xlsx
    try:
        import openpyxl
    except ImportError:
        print("Для сохранения в .xlsx необходимо установить библиотеку openpyxl.")
        print("Выполните команду: pip install openpyxl")
        return pd.DataFrame()

    # 1. Получаем и фильтруем список файлов масок
    try:
        mask_files = sorted(
            [f for f in os.listdir(mask_dir) if f.endswith('_cp_masks.png')], 
            key=lambda x: int(x.split('_')[1])
        )
    except IndexError:
        raise ValueError("Формат имен файлов не соответствует 'prefix_framenumber_...'. Не удалось извлечь номер кадра.")
        
    if not mask_files:
        raise FileNotFoundError(f"Не найдены файлы масок с '_cp_masks.png' в директории: {mask_dir}")
        
    all_objects_data = []

    print(f"Анализируем {len(mask_files)} масок...")

    # 2. Итерируемся по всем маскам
    for mask_file in tqdm(mask_files, desc="Анализ масок"):
        frame_number = int(mask_file.split('_')[1])
        
        # Загружаем маску и соответствующее ей изображение
        mask_path = os.path.join(mask_dir, mask_file)
        image_file = mask_file.replace('_cp_masks.png', '.png')
        image_path = os.path.join(mask_dir, image_file)

        try:
            mask = imread(mask_path)
            # Загружаем оригинальное изображение как grayscale для анализа интенсивности
            image = imread(image_path, as_gray=True)
        except FileNotFoundError as e:
            print(f"Предупреждение: Пропущен кадр {frame_number}, так как файл не найден: {e}")
            continue

        # 3. Извлекаем свойства для каждого объекта на маске
        # `intensity_image` позволяет вычислить статистику по интенсивности пикселей
        regions = regionprops(mask, intensity_image=image)
        
        for region in regions:
            obj_data = {'frame': frame_number, 'label_in_frame': region.label}
            
            # Извлекаем стандартные свойства
            for prop in properties_to_measure:
                obj_data[prop] = getattr(region, prop)
            
            # 'intensity_std' не является стандартным свойством, вычисляем его вручную
            # region.intensity_image содержит значения пикселей оригинального изображения под маской объекта
            obj_data['intensity_std'] = np.std(region.intensity_image)
            
            all_objects_data.append(obj_data)

    if not all_objects_data:
        print("Не найдено ни одного объекта в масках. DataFrame не будет создан.")
        return pd.DataFrame()

    # 4. Создаем DataFrame
    df = pd.DataFrame(all_objects_data)
    
    # 5. Если указан размер пикселя, добавляем колонки в мкм
    if pixel_size_um is not None:
        print(f"Пересчитываем размеры с учетом размера пикселя: {pixel_size_um} мкм/пикс")
        # Площадь в кв. мкм
        df['area_um2'] = df['area'] * (pixel_size_um ** 2)
        # Длина в мкм
        df['axis_major_length_um'] = df['axis_major_length'] * pixel_size_um
        # Переименовываем оригинальные колонки для ясности
        df = df.rename(columns={'area': 'area_pixels', 'axis_major_length': 'axis_major_length_pixels'})
        # Упорядочиваем колонки для удобства
        desired_order = [
            'frame', 'label_in_frame', 'area_pixels', 'area_um2', 
            'axis_major_length_pixels', 'axis_major_length_um',
            'solidity', 'eccentricity', 'mean_intensity', 'intensity_std'
        ]
        # Убедимся, что все колонки присутствуют, прежде чем переупорядочивать
        final_columns = [col for col in desired_order if col in df.columns]
        df = df[final_columns]


    # 6. Отображаем первые 5 строк в консоли Jupyter Notebook
    print("\nПервые 5 строк полученной таблицы:")
    display(df.head())

    # 7. Сохраняем DataFrame в Excel
    try:
        df.to_excel(output_path, index=False, engine='openpyxl')
        print(f"\nТаблица успешно сохранена в: {output_path}")
    except Exception as e:
        print(f"\nОшибка при сохранении файла: {e}")
        
    return df
```
